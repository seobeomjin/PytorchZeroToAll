```python
import torch 
from torch.autograd import Variable 
import torch.nn.functional as F 

x_data = Variable(torch.Tensor([[1.0],[2.0],[3.0],[4.0]]))  # x data set 
y_data = Variable(torch.Tensor([[0.],[0.],[1.],[1.]])) # y data set 
# y data is little different because logistic regression which is binary 

# 1) design the model
class Model(torch.nn.Module):
    
    def __init__(self):
        """
        In the constructor we instantiate nn.Linear module
        """
        super(Model,self).__init__()
        self.linear = torch.nn.Linear(1,1) # One in and one out 
        
    def forward(self,x):
        #y_pred = F.sigmoid(self.linear(x))
        
        """
        In the forward function we accept a Variable of input data and 
        we must return a Variable of output data. 
            
        """
        y_pred = F.sigmoid(self.linear(x))
        # predictive value y  
        # use sigmoid function as activation function
        return y_pred
        # return forward propagation value 
        

# our model        
model = Model()


# 2) Construct our loss function and an Optimizer.

# Construct our loss function and an Optimizer. The call to model.parameters()
# in the SGD constructor will contain the learnable parameters of the two
# nn.Linear modules which are members of the model.

criterion = torch.nn.BCELoss(size_average=True)
# Use Binary Cross Entropy as loss function 
optimizer = torch.optim.SGD(model.parameters(),lr = 0.01)
# Use Stochastic Gradient Descent for updating parameter 
# model.parameter() let function know the parameter which we need to be updated

# Training loop
for epoch in range (500):
        # Forward pass: Compute predicted y by passing x to the model

    y_pred = model(x_data)
    
    # Compute and print loss 
    
    loss = criterion(y_pred,y_data)
    #put the loss value in variavble loss  
    print(epoch,loss.data)
    # print loss data for each epoch 
    
    # Zero gradients, perform a backward pass, and update the weights.
    
    optimizer.zero_grad()
    # initilaize for each epoch
    loss.backward()
    # loss back propagation
    optimizer.step()
    # updating variables 
    
    
# After training
hour_var = Variable(torch.Tensor([[1.0]]))  # get value one 
print("predict 1 hour ", 1.0, model(hour_var).data[0][0] > 0.5)  #pass this value to the model 
hour_var = Variable(torch.Tensor([[7.0]]))
print("predict 7 hours", 7.0, model(hour_var).data[0][0] > 0.5)
# checking after training whether true or false 




        
        
        
        
```
