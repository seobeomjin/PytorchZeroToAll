```python
from torch import tensor 
from torch import nn
import torch.optim as optim
####
from torch import sigmoid 
import torch.nn.functional as F  ## it was for F.sigmoid(self.linear(x))

"""
### ISSUE ### 
Is there any differnce between torch.sigmoid and torch.nn.functional.sigmoid ? 

Actually, there is a difference.
The implementations in torch.xxx have the backward implemented using python calls, 
while the functional counterparts have their backward implemented entirely in C/Cuda, 
so the functional backward code is more efficient.

source : https://discuss.pytorch.org/t/is-there-any-different-between-torch-sigmoid-and-torch-nn-functional-sigmoid/995
"""
####


#Training data and ground truth 
x_data = tensor([[1.0],[2.0],[3.0],[4.0]])
y_data = tensor([[0.],[0.],[1.],[1.]])

class Model(nn.Module):
    def __init__(self):
        """
        constructor of the model , we instantiate nn.Linear Module
        """
        super(Model,self).__init__()
        self.linear = nn.Linear(1,1) # One in and One out 
        
    def forward(self,x):
        """
        accept a Variable of input data and return a Varaible of output data
        """
        y_pred = sigmoid(self.linear(x)) 
        # using sigmoid as activation function 
        # F.sigmoid(self.linear(x)) -> no matter,too 
        return y_pred
    
#our model 
model = Model()

criterion = nn.BCELoss(reduction='mean')
# Use Binary Cross Entropy as loss function 

optim = optim.SGD(model.parameters(),lr=0.01)

#Training loop
for epoch in range(1000):
    # 1) forward pass 
    y_pred = model(x_data)
    
    # 2) compute gradient and print 
    loss = criterion(y_pred,y_data) 
    print(f'Epoch {epoch +1}/1000 | Loss: {loss.item():.4f}')  
    # '  :.4f' 는 콜론 앞의 변수에 대해 콜론 뒤의 형식으로 적어주는 것을 의미
    # 여기서는 앞의 변수에 대해 소수점 넷째 자리까지만을 표기함을 의미 
    
    # 3) gradient cleaning, calculate gradient, updata the parameters 
    optim.zero_grad()
    loss.backward()
    optim.step()
    
#After training 
print(f'\nLet\'s predict the hours need to score above 50%\n{"="*50}')
#상위 50%를 넘기기 위해 학습에 필요한 시간(?) 

hour_list = []
percentage_list = []

hour_var = model(tensor([[1.0]]))
print(f'Prediction after 1 hour of training: {hour_var.item():.4f} | Above 50% : {hour_var.item() > 0.5}')
hour_var = model(tensor([[1.5]]))
print(f'Prediction after 1.5 hours of training: {hour_var.item():.4f} | Above 50%: {hour_var.item() > 0.5}')
hour_var = model(tensor([[2.0]]))
print(f'Prediction after 2 hours of training: {hour_var.item():.4f} | Above 50%: {hour_var.item() > 0.5}')
hour_var = model(tensor([[3.0]]))
print(f'Prediction after 3 hours of training: {hour_var.item():.4f} | Above 50%: {hour_var.item() > 0.5}')
hour_var = model(tensor([[6.0]]))
print(f'Prediction after 6 hours of training: {hour_var.item():.4f} | Above 50%: {hour_var.item() > 0.5}')

    
        
```

```python
#previous version (2019)

import torch 
from torch.autograd import Variable 
import torch.nn.functional as F 

x_data = Variable(torch.Tensor([[1.0],[2.0],[3.0],[4.0]]))  # x data set 
y_data = Variable(torch.Tensor([[0.],[0.],[1.],[1.]])) # y data set 
# y data is little different because logistic regression which is binary 

# 1) design the model
class Model(torch.nn.Module):
    
    def __init__(self):
        """
        In the constructor we instantiate nn.Linear module
        """
        super(Model,self).__init__()
        self.linear = torch.nn.Linear(1,1) # One in and one out 
        
    def forward(self,x):
        #y_pred = F.sigmoid(self.linear(x))
        
        """
        In the forward function we accept a Variable of input data and 
        we must return a Variable of output data. 
            
        """
        y_pred = F.sigmoid(self.linear(x))
        # predictive value y  
        # use sigmoid function as activation function
        return y_pred
        # return forward propagation value 
        

# our model        
model = Model()


# 2) Construct our loss function and an Optimizer.

# Construct our loss function and an Optimizer. The call to model.parameters()
# in the SGD constructor will contain the learnable parameters of the two
# nn.Linear modules which are members of the model.

criterion = torch.nn.BCELoss(size_average=True)
# Use Binary Cross Entropy as loss function 
optimizer = torch.optim.SGD(model.parameters(),lr = 0.01)
# Use Stochastic Gradient Descent for updating parameter 
# model.parameter() let function know the parameter which we need to be updated

# Training loop
for epoch in range (500):
        # Forward pass: Compute predicted y by passing x to the model

    y_pred = model(x_data)
    
    # Compute and print loss 
    
    loss = criterion(y_pred,y_data)
    #put the loss value in variavble loss  
    print(epoch,loss.data)
    # print loss data for each epoch 
    
    # Zero gradients, perform a backward pass, and update the weights.
    
    optimizer.zero_grad()
    # initilaize for each epoch
    loss.backward()
    # loss back propagation
    optimizer.step()
    # updating variables 
    
    
# After training
hour_var = Variable(torch.Tensor([[1.0]]))  # get value one 
print("predict 1 hour ", 1.0, model(hour_var).data[0][0] > 0.5)  #pass this value to the model 
hour_var = Variable(torch.Tensor([[7.0]]))
print("predict 7 hours", 7.0, model(hour_var).data[0][0] > 0.5)
# checking after training whether true or false 




        
        
        
        
```
