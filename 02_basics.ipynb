```python
import numpy as np 
import matplotlib.pyplot as plt 

# 실제 보유하고 있는 데이터셋 설정 
x_data = [1.0,2.0,3.0]   
y_data = [3.0,5.0,7.0] 

#out model for the forward pass 
# weight와 x를 곱하여 y값을 예측하는 함수이다.
def forward(x):
    return x*w

#Loss function 
#MSE를 구하는 과정 중에 있으며, 예측된 y값에서 
# 실제 y값을 빼낸 뒤 제곱한다.  
def loss(x,y):
    y_pred = forward(x)
    loss_ing = ((y_pred)-y)*((y_pred)-y)
    return loss_ing

#weight 와 mse 값을 받을 list를 생성
w_list = []
mse_list =[]

#w를 0.0부터 4.0까지 0.1씩 증가 
for w in np.arange(0.0,4.0,0.1):   
    print("w=",w) #w 출력 
    I_sum = 0 # Loss sum 이 들어간다.
    #x와 y 각각의 데이터셋에서 하나씩 꺼낸다. 
    for x_val,y_val in zip(x_data,y_data):
        y_pred_val = forward(x_val) #y 예상값
        I = loss(x_val,y_val) # 각 경우 loss 계산
        I_sum += I # 데이터셋의 데이터 갯수만큼 오차도 더해진다. 
        print('\t',x_val,y_val,y_pred_val,I) 
        # x,y 데이터, y 데이터 예측값, 오차값 출력 
    print("MSE=",I_sum/3) #MSE 출력 
    #list에 값을 넣는다. 
    w_list.append(w) 
    mse_list.append(I_sum/3)
```

```python
#그래프를 설정하고 축에 명시를 하고 그래프를 보인다. 
plt.plot(w_list,mse_list)
plt.ylabel('Loss')
plt.xlabel('w')
plt.show()
    
```
